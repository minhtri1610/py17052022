```python
df_maths.loc[df['TBDH'] < 5, 'KETQUA'] = 'FAIL'
df_maths.loc[df['TBDH'] > 5, 'KETQUA'] = 'PASS'

df_maths
```

```python
df_dh = df_maths

df_dh.groupby(['KV'])['KV'].agg(['count'])

df_dh.groupby(['KV'])['DH1', 'DH2', 'DH3'].agg(['min', 'mean', 'max', 'std'])
#std cot DH1 cua KV 2 >  KV2NT  =>  du lieu it bien dong hon
# -> chất lượng giáo dục của kv2 đồng đều hơn
#df_dh
```

```python
pd.pivot_table(df_dh, values=['DH1', 'DH2', 'DH3'], columns='KETQUA', aggfunc=['min', 'max', 'mean'])
```
